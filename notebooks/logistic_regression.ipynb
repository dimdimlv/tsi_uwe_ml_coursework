```python
import matplotlib.pyplot as plt
import numpy as np

# Extract metrics for plotting
experiment_ids = [1, 2, 3, 4, 5, 6]  # Replace with actual Experiment IDs
accuracy = [73.84, 69.87, 76.14, 66.27, 69.30, 66.60]  # Replace with actual Accuracy values
f1_scores = [0.512 * 100, 0.674 * 100, 0.517 * 100, 0.706 * 100, 0.675 * 100, 0.705 * 100]  # Convert F1-Score to %
roc_auc = [0.743 * 100, 0.751 * 100, 0.743 * 100, 0.752 * 100, 0.749 * 100, 0.750 * 100]  # Convert ROC-AUC to %


# Bar chart configuration
x = np.arange(len(experiment_ids))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Plot each metric
rects1 = ax.bar(x - width, accuracy, width, label='Accuracy', color='blue')
rects2 = ax.bar(x, f1_scores, width, label='F1-Score (Default)', color='orange')
rects3 = ax.bar(x + width, roc_auc, width, label='ROC-AUC', color='green')

# Add labels, title, and legend
ax.set_xlabel('Experiment ID')
ax.set_ylabel('Scores')
ax.set_title('Comparison of Accuracy, F1-Score, and ROC-AUC Across Experiments')
ax.set_xticks(x)
ax.set_xticklabels(experiment_ids)
ax.legend()

# Annotate bar heights
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

# Show plot
plt.tight_layout()
plt.show()
```
